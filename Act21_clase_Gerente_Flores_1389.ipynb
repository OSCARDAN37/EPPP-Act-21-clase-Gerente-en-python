```python
class Gerente:
    def __init__(self, nombre, telefono, edad):
        self.nombre = nombre
        self.telefono = telefono
        self.edad = edad

    def supervisar(self):
        print(f"El gerente {self.nombre} está supervisando la productividad del equipo.")

    def planificar(self):
        print(f"El gerente {self.nombre} está planificando las metas del próximo trimestre.")

    def firmar_contrato(self):
        print(f"El gerente {self.nombre} ha firmado un nuevo contrato comercial.")

gerente1 = Gerente("Carlos Ruiz", "555-0101", 42)

gerente2 = Gerente("Ana Martínez", "555-0202", 35)

print("Información del Gerente 1:")
print(f"Nombre: {gerente1.nombre}, Tel: {gerente1.telefono}, Edad: {gerente1.edad}")
gerente1.supervisar()
gerente1.planificar()

print("\n" + "-"*30 + "\n")

print("Información del Gerente 2:")
print(f"Nombre: {gerente2.nombre}, Tel: {gerente2.telefono}, Edad: {gerente2.edad}")
gerente2.firmar_contrato()
gerente2.supervisar()

print("\nPrograma terminado")
print("Creado de software pro Oscar Flores NL = 19")
```
